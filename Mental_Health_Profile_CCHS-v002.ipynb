```r
library(dplyr)
library(readr)
library(ggplot2)
library(ggthemes)
library(psych)
```

```r
df <- read_csv("https://raw.githubusercontent.com/tarjin/Canada_Open_Data/master/Mental_Health_CCHS_01051101-eng.csv")
```

```r
glimpse(df)
```

```r
names(df) <- c("Year", "Geo", "Age_Group", "Sex", "Indicator", "Characteristic", "Vector", "Coordinate", "Value")
```

```r
head(df)
```

```r
tail(df)
```

```r
df$Geo <- factor(df$Geo)
```

```r
df$Age_Group <- factor(df$Age_Group)
```

```r
df$Sex <- factor(df$Sex)
```

```r
df$Indicator <- factor(df$Indicator)
```

```r
df$Characteristic <- factor(df$Characteristic)
```

```r
glimpse(df)
```

```r
describe(df) # from psych package
```

```r
df_Suicidal_thoughts_12_months <- df %>%
  filter(Year == 2012) %>%
  filter(Characteristic == "Number of persons") %>%
  filter(Age_Group == "Total, 15 years and over") %>%
  filter(Sex %in% c("Females", "Males")) %>%
  filter(Geo %in% c("Nova Scotia", "Ontario", 
                    "Alberta", "British Columbia" )) %>%
  filter(Indicator == "Suicidal thoughts, 12 months") 
```

```r
glimpse(df_Suicidal_thoughts_12_months)
```

```r
df_Suicidal_thoughts_12_months$Value <- as.integer(df_Suicidal_thoughts_12_months$Value)
```

```r
glimpse(df_Suicidal_thoughts_12_months)
```

```r
myplot <- ggplot(data = df_Suicidal_thoughts_12_months) +
  geom_col(aes(x = Geo, y = Value, fill = Sex)) +
  ylab("Number of Persons") +
  xlab(" ") +
  theme(axis.text.x = element_text(angle = 60, hjust = 1)) +
  ggtitle("Number of Persons with Suicidal Thoughts 2012\n(within last 12 months by Sex)\nfor NS, ON, AB, and BC") +
  theme(plot.title = element_text(hjust = 0.5)) +
  scale_y_continuous(labels=function(n){format(n, scientific = FALSE)}) +
  scale_fill_grey(start = 0, end = 0.5) +
  coord_flip()
```

```r
myplot
```
