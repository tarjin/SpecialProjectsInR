```r
library(dplyr)
library(data.table)
library(readr)
library(ggplot2)
library(ggthemes)
library(psych) 
```

```r
df <- fread("https://raw.githubusercontent.com/tarjin/Canada_Open_Data/master/Mental_Health_CCHS_01051101-eng.csv")
```

```r
df_Filtered_Indicator <- df %>%
  filter(Year == 2012) %>%
  filter(Characteristic == "Percent") %>%
  filter(Sex %in% c("Males")) %>%
  filter(Age_Group %in% c("45 to 64 years")) %>%
  filter(Geo %in% c("Nova Scotia")) %>%
  filter(Value >= 0)
```

```r
glimpse(df_Filtered_Indicator)
```

```r
df_Filtered_Indicator <- df_Filtered_Indicator %>%
  filter(!Value %in% c("F", "M"))
```

```r
glimpse(df_Filtered_Indicator)
```

```r
df_Filtered_Indicator$Value <- round(as.numeric(df_Filtered_Indicator$Value))
```

```r
glimpse(df_Filtered_Indicator)
```

```r
myplot <- ggplot(data = df_Filtered_Indicator) +
  geom_col(aes(x = reorder(Indicator, Value), y = Value), fill = "cadetblue4") +
  ylab("                                 Percent of Males 45-64 Years") +
  theme(axis.text.y = element_text(angle = 0, hjust = 0, size = 9)) +
  theme(axis.title.y = element_text(angle = 0, hjust = 0, size = 9)) +
  xlab(" ") +
  theme(axis.text.x = element_text(angle = 0, hjust = 0, size = 9)) +
  theme(axis.title.x = element_text(angle = 0, hjust = 0, size = 9)) +
  ggtitle("Percent (> 0%) of Males 45-64 Years\nReporting Each Mental Health Measure\nfor Nova Scotia, 2012") +
  theme(plot.title = element_text(hjust = 0.5, size = 12)) +
  scale_y_continuous(labels=function(n){format(n, scientific = FALSE)}) +
  coord_flip()
```
