```r
library(dplyr)
library(data.table)
library(readr)
library(ggplot2)
library(ggthemes)
library(psych)
```

```r
df <- fread("https://raw.githubusercontent.com/tarjin/Canada_Open_Data/master/Mental_Health_CCHS_01051101-eng.csv")
```

```r
glimpse(df)
```

```r
names(df) <- c("Year", "Geo", "Age_Group", "Sex", "Indicator", "Characteristic", "Vector", "Coordinate", "Value")
```

```r
head(df)
```

```r
tail(df)
```

```r
df$Geo <- factor(df$Geo)
```

```r
df$Age_Group <- factor(df$Age_Group)
```

```r
df$Sex <- factor(df$Sex)
```

```r
df$Indicator <- factor(df$Indicator)
```

```r
df$Characteristic <- factor(df$Characteristic)
```

```r
glimpse(df)
```

```r
describe(df) # from psych package
```

```r
glimpse(df$Indicator)
```

```r
levels(factor(df$Indicator)) 
```

```r
df_Filtered_Indicator <- df %>%
  filter(Year == 2012) %>%
  filter(Characteristic == "Percent") %>%
  filter(Age_Group == "Total, 15 years and over") %>%
  filter(Sex %in% c("Females", "Males")) %>%
  filter(Geo %in% c("Newfoundland and Labrador", "Prince Edward Island", "Nova Scotia", 
                    "New Brunswick")) %>%
  filter(Indicator == "Self-rated work stress, days quite a bit or extremely stressful") 
```

```r
glimpse(df_Filtered_Indicator)
```

```r
df_Filtered_Indicator$Value <- as.integer(df_Filtered_Indicator$Value)
```

```r
glimpse(df_Filtered_Indicator)
```

```r
myplot <- ggplot(data = df_Filtered_Indicator) +
  geom_col(aes(x = Geo, y = Value, fill = Sex)) +
  ylab("                Percent of Persons") +
  theme(axis.text.y = element_text(angle = 0, hjust = 0, size = 9)) +
  theme(axis.title.y = element_text(angle = 0, hjust = 0, size = 9)) +
  xlab(" ") +
  theme(axis.text.x = element_text(angle = 0, hjust = 0, size = 9)) +
  theme(axis.title.x = element_text(angle = 0, hjust = 0, size = 9)) +
  ggtitle("Percent of Persons Reporting\nExtreme Work Stress\nfor Atlantic Provinces, 2012") +
  theme(plot.title = element_text(hjust = 0.5, size = 9)) +
  scale_y_continuous(labels=function(n){format(n, scientific = FALSE)}) +
  scale_fill_grey(start = 0, end = 0.5) +
  coord_flip()
```

```r
myplot
```
